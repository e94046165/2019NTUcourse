```python
import numpy as np

theta = np.radians(45)

r = np.array(( (np.cos(theta), -np.sin(theta)),
               (np.sin(theta),  np.cos(theta)) ))

print('rotation matrix:')
print(r)

v = np.array((0,1))

print('vector v: ')
print(v)

print('apply the rotation matrix r to v: r*v')
print( r.dot(v) )
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('lena.bmp')
# pyplot  Img  jupyter notebook 
def imshow(img):
    plt.imshow(img[:,:,::-1])# plt  cv  RGB  BGR
imshow(img)

def rotation(img, degree):
    newimg = np.zeros((len(img),len(img[0]),3), np.uint8)
    cent_x = len(img[0])/2
    cent_y = len(img)/2
    theta = np.radians(degree)
    r = np.array(( (np.cos(theta), -np.sin(theta)),(np.sin(theta),  np.cos(theta))))
    for i in range(len(img)):
        for j in range(len(img[0])):
            v = np.array((i - cent_y, j - cent_x))
            v = r.dot(v)
            v = (int(round(v[0]+cent_y)), int(round(v[1]+cent_x)))
            if v[0] > len(img)-1 or v[1]>len(img)-1 or v[0]<0 or v[1]<0:
                pass
            else:
                newimg[v[0]][v[1]] = img[i][j]
    return newimg
imshow(rotation(img, -45))
```
